```python
# importing necessary modules to complete project.
from pymongo import MongoClient
import pandas as pd
import json
```

```python
# setting up a connection to a mongodb.
client = MongoClient('localhost', 27017)
```

```python
# creating a new database.
mongo_db = client['economic_data_db']
```

```python
# creating a new collection within the new database.
gdp_collection = mongo_db['gdp_growth']
unemployment_collection = mongo_db['unemployment_growth']
```

```python
# reading in the csv file using pandas, skiping unnecessary rows, and naming columns by years.
gdp_df = pd.read_csv('Resources/gdp_countries.csv', skiprows=[0,4], names=["country_name", "country_abb", "data_type", "info", "1960","1961","1962","1963","1964","1965","1966","1967","1968","1969","1970","1971","1972","1973","1974","1975","1976","1977","1978","1979","1980","1981","1982","1983","1984","1985","1986","1987","1988","1989","1990","1991","1992","1993","1994","1995","1996","1997","1998","1999","2000","2001","2002","2003","2004","2005","2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017","2018"])
unemployment_df = pd.read_csv('Resources/unemployment_data.csv', skiprows=[0,4], names=["country_name", "country_abb", "data_type", "info", "1960","1961","1962","1963","1964","1965","1966","1967","1968","1969","1970","1971","1972","1973","1974","1975","1976","1977","1978","1979","1980","1981","1982","1983","1984","1985","1986","1987","1988","1989","1990","1991","1992","1993","1994","1995","1996","1997","1998","1999","2000","2001","2002","2003","2004","2005","2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017","2018"])
```

```python
# dropping first row that contains unnecessary information.
gdp_df = gdp_df.drop(gdp_df.index[0])
unemployment_df = unemployment_df.drop(gdp_df.index[0])
```

```python
# dropping unnecessary column.
gdp_df = gdp_df.drop(columns=['info', '1960'])
unemployment_df = unemployment_df.drop(columns=['info', '1960'])
```

```python
# setting the index of the dataframe to country name.
gdp_df = gdp_df.set_index('country_name')
unemployment_df = unemployment_df.set_index('country_name')
```

```python
# making a dataframe of a select few countries.
gdp_df = gdp_df.loc[['United States', 'China', 'Russian Federation', 'Brazil', 'Germany', 'India', 'Singapore', 'Japan', 'Kenya', 'Mexico', 'Panama', 'Australia', 'Netherlands', 'Norway', 'Finland', 'Sweden', 'Switzerland', 'France', 'Spain', 'United Kingdom', 'Indonesia', 'Argentina', 'South Africa', 'Colombia', 'Denmark', 'Iceland', 'United Arab Emirates', 'Saudi Arabia']]
unemployment_df = unemployment_df.loc[['United States', 'China', 'Russian Federation', 'Brazil', 'Germany', 'India', 'Singapore', 'Japan', 'Kenya', 'Mexico', 'Panama', 'Australia', 'Netherlands', 'Norway', 'Finland', 'Sweden', 'Switzerland', 'France', 'Spain', 'United Kingdom', 'Indonesia', 'Argentina', 'South Africa', 'Colombia', 'Denmark', 'Iceland', 'United Arab Emirates', 'Saudi Arabia']]
```

```python
# reseting the index of the dataframe for purposes.
gdp_df = gdp_df.reset_index()
unemployment_df = unemployment_df.reset_index()
```

```python
# turning above refined pandas dataframe into json format.
gdp_data_json = json.loads(gdp_df.to_json(orient='records'))
unemployment_data_json = json.loads(unemployment_df.to_json(orient='records'))
```

```python
# clearing the mongodatabase.
gdp_collection.remove()
unemployment_collection.remove()
```

```python
# adding updated information to mongodatabase.
gdp_collection.insert(gdp_data_json)
unemployment_collection.insert(unemployment_data_json)
```
